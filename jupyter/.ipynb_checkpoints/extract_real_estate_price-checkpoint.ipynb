```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from bs4 import BeautifulSoup
import json
from hdfs import InsecureClient
client = InsecureClient('http://localhost:9870', user='big')
```

```python
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument("--single-process")
chrome_options.add_argument("--disable-dev-shm-usage")
s = Service('/home/big/study/spark/driver/chromedriver')
driver = webdriver.Chrome(service=s, options=chrome_options)
```

```python
driver.get('https://data.kbland.kr/kbstats/wmh?tIdx=HT01&tsIdx=monthAptSalePriceInx')
```

```python
xpath_10 = '//*[@id="container"]/div/div/div[1]/section/div[2]/div[2]/div/div/div[2]/div/div/div[2]/label[2]'
driver.find_element(By.XPATH, xpath_10).click()
```

```python
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
soup
```

```python
# 날짜 데이터 가공
bs_dayitems = soup.findAll('div', {'class': 'dayitem'})
dayitems = []
for item in bs_dayitems[16:]:
    dayitems.append(item.text.replace('.', '-'))
print(dayitems, len(dayitems))
```

```python
cols = ['loc', 'std_date', 'trading_price_index']
loc_items = [('전국', 0), ('서울', 1), ('강북14개구', 2), ('강남11개구', 3), ('수도권', 4), ('6개광역시', 5), ('5개광역시', 6), ('기타지방', 7),  
             ('부산', 8), ('대구', 9), ('인천', 10), ('광주', 11), ('대전', 12), ('울산', 13), ('세종', 14), 
             ('경기', 15), ('강원', 16), ('충북', 17), ('충남', 18), ('전북', 19), ('전남', 20), ('경북', 21), ('경남', 22), ('제주', 23)]
data = []
for item in loc_items:
    class_name = 'bodyitem hover_' + str(item[1])
    bs_price_items = soup.findAll('div', {'class': class_name})
    bs_price_items = bs_price_items[17:]
    for i in range(len(bs_price_items)):
        rows = [item[0], dayitems[i], float(bs_price_items[i].text)]
        data.append(dict(zip(cols, rows)))
print(data)
```

```python
# json 형태로 만듦
res_apart = {
    'meta':{
        'desc': '월간 아파트 매매가격지수',
        'cols': {
            'loc': '지역명',
            'std_date': '기준 년월',
            'trading_price_index': '매매가격지수',
        }
    },
    'data': data
}
res_apart
```

```python
FILE_DIR = '/real_estate/price_index/apartment/'
FILE_NAME = 'price_index_apart_' + dayitems[0] + '_' + dayitems[-1] + '.json'
res_apart_json = json.dumps(res_apart, ensure_ascii=False)
client.write(FILE_DIR + FILE_NAME, res_apart_json, overwrite=True, encoding='utf-8') # hdfs에 저장
```

```python
driver.close()
```

```python
import datetime as dt
```

```python
def cal_std_month(before_month):
    x = dt.datetime.now() - dt.timedelta(30*before_month) #현재날짜-지정날짜
    year = x.year #연도
    month = x.month if x.month >= 10 else '0'+ str(x.month) #월
    return str(year)+ '-' +str(month)
```

```python
for i in range(0, 30):
    print(cal_std_month(i))
```
